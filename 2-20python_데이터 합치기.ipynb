```python
import pandas as pd
```

# concatÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ Ìï©ÏπòÍ∏∞

```python
df1 = pd.DataFrame({'A': [1,3,5,7,9],
                    'B': [11,12,13,14,15],
                    'C': [0,1,2,3,4],
                    'D': [101,102,103,104,105]},
                 index = [1,2,3,4,5] )
df1
```

```python
df2 = pd.DataFrame({'A': [2,4,6,8,10],
                    'B': [1,2,3,4,5],
                    'C': [30,31,32,33,34],
                    'D': [1011,1021,1031,1041,1051]    
},
                  index = [6,7,8,9,10])
df2
```

```python
#concat(Ïª®Ï∫£) ÏúÑÏïÑÎûòÎ°ú Î∂ôÏù¥Í∏∞ (Í∏∞Î≥∏Í∞í)
pd.concat([df1,df2])
```

```python
#Î®ºÏ†Ä Ïù∏Îç±Ïä§ ÌÜµÏùº
# reset_index : Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò Ïù∏Îç±Ïä§Î•º 0,1,2,3... Î°ú Î∞îÍøîÏ§ÄÎã§. Ïù∏Îç±Ïä§Î•º 0Î∂ÄÌÑ∞ Îã§Ïãú ÏÉàÎ°ú ÎßåÎìúÎäî Ìï®ÏàòÏïº.
df1.reset_index()
```

```python
 #ÏïàÏóêÏûàÎäî Ïù∏Îç±Ïä§ Ïª¨Îüº ÎìúÎ°≠ÏúºÎ°ú ÏÇ≠Ï†ú
df1.reset_index(drop=True) 
```

```python
#üîπ Í∏∞Î≥∏ Í∞úÎÖê‚úÖ 1Ô∏è‚É£ inplace=False (Í∏∞Î≥∏Í∞í)üëâ ÏõêÎ≥∏ÏùÄ Í∑∏ÎåÄÎ°ú ÎëêÍ≥†, ÏÉàÎ°úÏö¥ Í≤∞Í≥ºÎ•º Î∞òÌôò
#‚úÖ 2Ô∏è‚É£ inplace=Trueüëâ ÏõêÎ≥∏ df ÏûêÏ≤¥Î•º Î∞îÎ°ú ÏàòÏ†ï
df1.reset_index(drop=True, inplace=True)
df1
```

```python
df2.reset_index(drop=True, inplace=True)
df2
```

```python
# concatÏúºÎ°ú Îç∞Ïù¥ÌÑ∞Î•º ÏñëÏòÜÏúºÎ°ú Ìï©ÏπòÎ†§Î©¥?? # axis=0 : Ìñâ Í¥ÄÎ†®, axis=1 : Ïó¥ Í¥ÄÎ†® (ÏóëÏãúÏä§)
# concatÏúºÎ°ú df1, df2 Îç∞Ïù¥ÌÑ∞ Ìï©ÏπòÍ∏∞
#ÌåêÎã§Ïä§ÏóêÏÑú axis ÏùòÎØ∏
#axis=0 ‚Üí Ìñâ Î∞©Ìñ• (ÏÑ∏Î°ú)
#axis=1 ‚Üí Ïó¥ Î∞©Ìñ• (Í∞ÄÎ°ú)
pd.concat([df1,df2], axis=1)
```

# mergeÎ°ú Îç∞Ïù¥ÌÑ∞ Ìï©ÏπòÍ∏∞

```python
df3 = pd.DataFrame({'id': ['X01','X02','X03','X04','X05'],
                    'A': [1,3,5,7,9],
                    'B': [11,12,13,14,15],
                    'C': [0,1,2,3,4],
                    'D': [101,102,103,104,105] })
df3
```

```python
df4 = pd.DataFrame({'id': ['X01','X02','X03','X04','X06'],
                    'E': [2,4,6,8,10],
                    'F': [1,2,3,4,5],
                    'G': [30,31,32,33,34],
                    'H': [1011,1021,1031,1041,1051] })
df4
```

```python
#merge (Î®∏ÏßÄ) Ìï©ÏπòÎã§,Î≥ëÌï©ÌïòÎã§
# on='id id Í∏∞Ï§ÄÏúºÎ°ú inner merge : Í≤πÏπòÎäî Í≤ÉÎßå Ìï©ÏπòÍ≤†Îã§ how='inner'(ÍµêÏßëÌï©)  
# df3, df4 ‚Üí Ìï©Ïπ† Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ
# on='id' ‚Üí id Ïª¨ÎüºÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Îß§Ïπ≠
# how='inner' ‚Üí ÏñëÏ™ΩÏóê Îã§ ÏûàÎäî Í∞íÎßå Ìè¨Ìï®

pd.merge(df3, df4, on='id', how='inner')
```

```python
#id Í∏∞Ï§ÄÏúºÎ°ú outer merge: Ï†ÑÎ∂Ä Îã§ Í∞ÄÏ†∏Ïò§Í≤†Îã§(Ìï©ÏßëÌï©) nan:Îç∞Ïù¥ÌÑ∞Í∞Ä ÏóÜÎã§ 
pd.merge(df3, df4, on='id', how='outer')
```

```python
# left merge
# ‚úî df3Ïóê ÏûàÎäî idÎäî Îã§ ÏÇ¥ÏïÑÏûàÏùå
# ‚úî df4ÏóêÎßå ÏûàÎäî id=4Îäî ÏÇ¨ÎùºÏßê
# ‚úî ÏóÜÎäî Í∞íÏùÄ NaN
# üìå Ï¶â: ÏôºÏ™Ω Í∏∞Ï§Ä Ïú†ÏßÄ
pd.merge(df3, df4, on='id', how='left')
```

```python
# right merge 
pd.merge(df3, df4, on='id', how='right')
```

```python
df5 = pd.DataFrame({'user_id': ['X01','X02','X03','X04','X06'],
                    'E': [2,4,6,8,10],
                    'F': [1,2,3,4,5],
                    'G': [30,31,32,33,34],
                    'H': [1011,1021,1031,1041,1051]    
})
df5
```

```python
# Í∏∞Ï§ÄÏù¥ ÎêòÎäî Ïª¨Îüº Ïù¥Î¶ÑÏù¥ ÏÑúÎ°ú Îã§Î•∏ Í≤ΩÏö∞
pd.merge(df3, df5, left_on='id', right_on='user_id')
```

```python
df1 = pd.DataFrame({'A': [1,3,5,7,9],
                    'B': [11,12,13,14,15],
                    'C': [0,1,2,3,4],
                    'D': [101,102,103,104,105]},
                  index = [1,2,3,4,5])
df1
```

```python
# index Í∏∞Ï§ÄÏúºÎ°ú joinÏù¥ ÏàòÌñâÎê®
df1.join(df2)
```
