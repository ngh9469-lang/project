# groupby

```python
#ÌçºÏ†∏ÏûàÎçò ÎèôÏùº ÏïÑÏù¥ÎîîÎ•º Ìï©Ï≥§Îã§ üëâ idÎ≥ÑÎ°ú Í∑∏Î£πÏùÑ Î¨∂Ïñ¥ÏÑú, Í∞Å Í∑∏Î£πÏùò Ïà´Ïûê Ïª¨ÎüºÏùÑ Î™®Îëê ÎçîÌïòÎùºÎäî ÎúªÏù¥Ïïº.
#üîπ 1Ô∏è‚É£ groupby('id') üëâ Í∞ôÏùÄ id Í∞íÏùÑ Í∞ÄÏßÑ ÌñâÎì§ÎÅºÎ¶¨ Î¨∂Ïñ¥Ï§å.
#üîπ 2Ô∏è‚É£ .sum()üëâ Í∞Å Í∑∏Î£π ÏïàÏùò Ïà´Ïûê Ïª¨ÎüºÏùÑ Î™®Îëê ÎçîÌï®
# ÏòàÏ†ú
# id	amount
# 1	   100
# 1	   200
# 2	   50
# 2	   70
# groupby('id') ÌïòÎ©¥
# id=1 Í∑∏Î£π ‚Üí 100, 200
# id=2 Í∑∏Î£π ‚Üí 50, 70
#Ïç∏ÏúºÎ°ú Ìï©Ï≥ê 1 300 2 120
    
df_trans.groupby('id').sum()
```

```python
# üëâ ÌèâÍ∑†(ÏÇ∞Ïà† ÌèâÍ∑†)ÏùÑ Íµ¨ÌïòÎäî Ìï®ÏàòÏïº. Í≥ÑÏÇ∞ Í≥ºÏ†ï(10 + 20 + 30) √∑ 3 = 20üëâ Í≤∞Í≥º: 20.0
df_trans.groupby('id').mean()
```

```python
#üëâ Îç∞Ïù¥ÌÑ∞Ïùò ÏöîÏïΩ ÌÜµÍ≥Ñ Ï†ïÎ≥¥Î•º Ìïú Î≤àÏóê Î≥¥Ïó¨Ï£ºÎäî Ìï®ÏàòÏïº.(ÎîîÏä§ÌÅ¨ÎùºÏù¥Î∏å)
df_trans.groupby('id').describe()
```
